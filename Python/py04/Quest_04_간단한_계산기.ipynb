```python
import math  # math 모듈을 임포트하여 제곱 연산에 사용

# 정수를 입력받는 함수
def get_integer_input(prompt):
    while True:
        try:
            value = int(input(prompt))  # 사용자로부터 정수를 입력 받음
            return value  # 정수가 입력되면 그 값을 반환
        except ValueError:
            # ValueError가 발생하면 잘못된 입력이므로 메시지를 출력하고 다시 입력받음
            print("잘못된 입력입니다. 정수를 입력해주세요.")

# 연산자를 입력받는 함수
def get_operator_input():
    while True:
        operator = input("연산자를 입력하세요 (+, -, *, /, **): ")  # 사용자로부터 연산자 입력 받기
        if operator in ['+', '-', '*', '/', '**']:  # 입력된 연산자가 지원되는 연산자인지 확인
            return operator  # 올바른 연산자일 경우 그 값을 반환
        else:
            # 지원되지 않는 연산자일 경우 오류 메시지를 출력하고 다시 입력 받음
            print("지원하지 않는 연산자입니다. +, -, *, /, **만 입력 가능합니다.")

# 계산을 수행하는 함수
def calculate(num1, num2, operator):
    try:
        if operator == '+':  # 덧셈 연산
            return num1 + num2
        elif operator == '-':  # 뺄셈 연산
            return num1 - num2
        elif operator == '*':  # 곱셈 연산
            return num1 * num2
        elif operator == '/':  # 나눗셈 연산
            if num2 == 0:  # 두 번째 숫자가 0이면 ZeroDivisionError 발생
                raise ZeroDivisionError("0으로 나눌 수 없습니다.")  # 0으로 나눌 수 없다는 오류 메시지 출력
            return num1 / num2  # 나눗셈 결과 반환
        elif operator == '**':  # 제곱 연산
            return math.pow(num1, num2)  # math.pow를 사용하여 제곱 계산
    except ZeroDivisionError as e:
        # ZeroDivisionError가 발생하면 오류 메시지 출력
        print(f"오류: {e}")
        return None  # 오류 발생 시 None을 반환

# 메인 계산기 로직
def main():
    while True:
        # 첫 번째 정수 입력 받기
        num1 = get_integer_input("첫 번째 정수를 입력하세요: ")

        # 연산자 입력 받기
        operator = get_operator_input()

        # 두 번째 정수 입력 받기
        num2 = get_integer_input("두 번째 정수를 입력하세요: ")

        # 계산 수행
        result = calculate(num1, num2, operator)

        if result is not None:  # 결과가 None이 아니라면 계산 결과 출력
            print(f"결과: {num1} {operator} {num2} = {result}")

        # 계속 계산할지 여부 확인
        continue_calculation = input("계속 계산하시겠습니까? (y/n): ")
        if continue_calculation.lower() != 'y':  # 'y'가 아니면 종료
            print("계산기를 종료합니다.")
            break  # 반복문 종료

# 프로그램 실행
main()

```
