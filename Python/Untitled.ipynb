```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))  #가로, 세로를 입력받는 코드
        area = width *  vertical # 넓이를 구하는 부분
        result = '직사각형의 넓이는 : ' + str(area)
        return  result

    def trape(self):
        base_width, top_width, vertical, = map(int, input('아랫변, 윗변, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))  #가로, 세로를 입력받는 코드
        area = (base_width + top_width) * vertical / 2 # 넓이를 구하는 부분
        result = '직사각형의 넓이는 : ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
# 평형사변형의 넓이
a.par()  # 3,2를 입력했을 때 넓이가 6이 나와야 합니다.
# 사다리꼴의 넓이
a.trape()  # 3,2,1 을 입력했을 때 넓이가 2.5이 나와야 합니다.
```

```python
# 문제 3-3. 클래스 업데이트  
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요! 

class Kiosk: 
    # 문제 2-1의 답을 입력하세요.     
    def __init__(self):
        self.menu = ["americano", "latte", "mocha", "yuza_tea", "green_tea", "choco_latte"]
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.price_sum = 0
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []  # 주문 리스트 초기화
        self.order_price = []  # 가격 리스트 초기화

        n = 0
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                self.price_sum += self.price[n-1]  # 선택 음료의 가격을 합산
                self.order_menu.append(self.menu[n-1])  # 주문한 메뉴 추가
                self.order_price.append(self.price[n-1])  # 해당 메뉴의 가격 추가
                print(self.menu[n-1], ' : ', self.price[n-1], '원')  # 메뉴 출력
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:  
                self.temp = 'HOT'
            elif t == 2: 
                self.temp = 'ICE'    
            else:    
                print("1과 2 중 하나를 입력하세요.\n")

        self.order_menu[-1] = self.temp + ' ' + self.menu[n-1]  # 온도 속성을 추가한 주문 결과를 출력

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서 
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if 1 <= n <= len(self.menu):
                # 추가 음료 온도 
                t = 0  # 기본값을 넣어주고
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:  
                        self.temp = 'HOT'
                    elif t == 2: 
                        self.temp = 'ICE'    
                    else:    
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(self.price[n-1])  
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.price_sum += self.price[n-1]
                print('추가 주문 음료', self.temp, self.menu[n-1], ':', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원')
            elif n == 0:
                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력합니다.
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불
    def pay(self):
        print(f'총 합계 금액은 {self.price_sum}원 입니다')
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:  # 1이나 2를 입력할 때까지 물어보기
            self.pay_method = int(input("현금은 1을, 카드는 2를 입력하세요 : "))
            if self.pay_method == 1:  
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2: 
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:    
                print('다시 결제를 시도해 주세요.')

    # 주문서 출력 
    def table(self):
        outline1 = '⟝' + '-' * 30 + '⟞'
        outline2 = '|' + ' ' * 31 + '|'

        print(outline1)
        for i in range(5):
            print(outline2)

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 : ', self.price_sum)

        for i in range(5):
            print(outline2)
        print(outline1)

        
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
